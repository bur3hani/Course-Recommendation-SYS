```python
import pickle
import numpy as np
import pandas as pd
import warnings

# Suppress sklearn warnings
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn")

# Load KNN Model
with open("/kaggle/input/course-recommender-system-models/scikitlearn/knn-based-collaborative-filtering/1/knn_model.pkl", "rb") as file:
    knn_model = pickle.load(file)

# Load NMF Model
with open("/kaggle/input/course-recommender-system-models/scikitlearn/knn-based-collaborative-filtering/1/nmf_model.pkl", "rb") as file:
    nmf_model = pickle.load(file)

# Load dataset for feature extraction (ensure correct path)
df = pd.read_csv("/kaggle/input/coursera-course-dataset/coursea_data.csv")

# Select the correct number of features expected by the model
knn_expected_features = knn_model.n_features_in_
example_course = df.select_dtypes(include=['number']).dropna().iloc[0, :knn_expected_features].values.reshape(1, -1)

# Generate recommendations using KNN
_, indices = knn_model.kneighbors(example_course)
print("Recommended Courses (KNN):", indices)

# Generate recommendations using NMF
nmf_recommendations = nmf_model.transform(example_course)
print("Recommended Courses (NMF):", nmf_recommendations)

# Save example notebook file
notebook_filename = "model_usage_demo.ipynb"
with open(notebook_filename, "w") as nb_file:
    nb_file.write("# Jupyter Notebook for Model Usage Demonstration\n")

print(f"Notebook saved: {notebook_filename}")
```
